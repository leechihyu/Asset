_substeps:_
>1. Change current directory to `~/auto-gpt` by running the command `cd auto-gpt`

>2. Run `docker run -it --env-file=.env -v $PWD:/app --rm autogpt --gpt3only`

>3. If the above command works, you will see:

>4. If you are in this case, press `n` to start your own task.

>5. Ask Auto-GPT to do whatever tasks you want (the command line can only display one line, so if your instruction is more than one line, the new lines will overwrite the first line automatically, but do not worry, it works well.)

>6. Auto-GPT will ask you for authorisation. You can input `y` for once-a-time authorisation or `y -n` for n-times authorisation. `y -10` means authorisation Auto-GPT to execute commands without asking for authorisation 10 times. Please avoid a large `n`, because the programme may get stuck in the middle and it will be extremely costly.



#### Quick Prompts
1. To paste content from you clipboard to the command prompt, <strong>right-click</strong> your mouse instead of pressing "Ctrl + V".
2. To exit the Auto-GPT programme, press "Ctrl + C". You can still use "Ctrl + C" to copy when the programme is not running.
3. To disconnect from the server, input "exit". Alternatively, you can simply close the window.
